```python
import zipfile
import os

# Path to your zip file
zip_file_path = '/content/DeepSpaceYoloDataset.zip'

# Extract the zip file to a specific directory
extract_dir = 'extracted_files/'

# Ensure the extraction directory exists
os.makedirs(extract_dir, exist_ok=True)

# Open the zip file and extract all its contents
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Files extracted to {extract_dir}")
```

```python
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import glob

# Path to the directory with your 300 random images
image_dir = '/content/extracted_files/DeepSpaceYoloDataset/images'

length = 100

# Load and preprocess images (resize to 32x32 for the generator)
def preprocess_images(image_dir, target_size=(32, 32)):
    image_paths = glob.glob(os.path.join(image_dir, '*.jpg'))  # Assuming images are in .jpg format
    random.shuffle(image_paths)

    images = []
    for img_path in image_paths[:length]:  # Select 300 random images
        img = load_img(img_path, target_size=target_size)  # Resize image
        img_array = img_to_array(img)  # Convert to numpy array
        img_array = (img_array / 127.5) - 1  # Normalize to [-1, 1]
        images.append(img_array)

    images = np.array(images)  # Convert to numpy array
    return images

# Data augmentation to increase the dataset to 1200 images
def augment_images(images, target_size=(32, 32), total_images=length*4):
    # Use ImageDataGenerator for augmentation
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    augmented_images = []
    i = 0

    for batch in datagen.flow(images, batch_size=32):
        for img in batch:
            img_resized = tf.image.resize(img, target_size)  # Resize to 64x64 for discriminator
            augmented_images.append(img_resized)

            i += 1
            if i >= total_images - len(images):
                break

        if i >= total_images - len(images):
            break

    augmented_images = np.array(augmented_images)
    augmented_images = np.concatenate((images, augmented_images[:total_images - len(images)]), axis=0)

    return augmented_images


    # Preprocess the original 300 images (resize to 32x32)
target_size = (64,64)
processed_images = preprocess_images(image_dir,target_size)

    # Augment the dataset to reach 1200 images (resize to 64x64 for discriminator)
augmented_images = augment_images(processed_images, target_size=target_size, total_images=length*4)

print(f"Original images shape: {processed_images.shape}")
print(f"Augmented images shape: {augmented_images.shape}")

```

```python
import os
import random
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Path to your images folder
image_dir = '/content/extracted_files/DeepSpaceYoloDataset/images'

# List all image files in the directory
image_files = os.listdir(image_dir)

# Randomly select 9 images
random_images = random.sample(image_files, 9)

# Plot the images in a 3x3 grid
plt.figure(figsize=(10, 10))
for i, img_file in enumerate(random_images):
    img_path = os.path.join(image_dir, img_file)

    # Load the image
    img = load_img(img_path)#, target_size=(64, 64))  # Resize if necessary
    img_array = img_to_array(img) / 255.0  # Normalize pixel values

    # Get original resolution of the image (before resizing)
    original_img = load_img(img_path)  # Load the original image without resizing
    original_width, original_height = original_img.size  # Get original resolution

    # Plot image with resolution as title
    plt.subplot(3, 3, i + 1)
    plt.imshow(img_array)
    plt.title(f'{original_width}x{original_height}', fontsize=12)  # Display resolution
    plt.axis('off')  # Hide axis

plt.tight_layout()  # Adjust layout so titles don’t overlap
plt.show()
```
